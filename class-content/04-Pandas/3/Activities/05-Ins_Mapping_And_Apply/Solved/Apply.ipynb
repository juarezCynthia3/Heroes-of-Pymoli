```python
import pandas as pd
```

```python
# Mapping lets you format an entire DataFrame
file = "Resources/sample_data.csv"
file_df = pd.read_csv(file)
file_df.head()
```

```python
def my_custom_functions(row):
    return row["avg_cost"] * row["population"]
file_df["Total Cost"] = file_df.apply(my_custom_functions, axis=1)
file_df.head()
```
